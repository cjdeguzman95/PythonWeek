```python
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
```

### Retrieving the information from the url and store in the "r" variable
#### The information is saved in a JSON format as a string datatype

```python
url = "https://api.exchangeratesapi.io/latest"
r = requests.get(url)
type(r.text)
```

```python
# This changes the value in "base" from EUR to USD
r = requests.get(url, params = {"base": "USD"})
print(r.text)
```

### Extracting just rates using the JSON package

```python
import json

# .loads() function converts the JSON strings into dictionary values
rates_json = json.loads(r.text)["rates"]
```

```python
# rates_json
# rates_json["GBP"]
gbp = float(rates_json["GBP"])
print("200USD = ", str(gbp*200), "GBP")
```

### Saving dictionary value in a variable

```python
sales = [(rates_json)]
```

### Creating a dataframe from a dictionary value

```python
df = pd.DataFrame(sales)
df
```

#### .transpose() swaps columns and rows

```python
df = df.transpose()
df
```

```python
# finding the index of the dataframe
df.index
```

```python
# this adds numerical indexes to the dataframe
df.reset_index(inplace = True)
df
```

```python
df = df.rename(columns = {"index": "country", 0: "exchange_rate"})
```
